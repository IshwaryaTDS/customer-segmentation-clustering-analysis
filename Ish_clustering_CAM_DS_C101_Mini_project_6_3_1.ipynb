Elbow method

```python
# Elbow method: tried different k.
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0, n_init = 10)
    kmeans.fit(small_seg)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

From Elbow method it looks like optimum number clusters would be 5. since the graph bend between 4 and 6.

```python
# Using our best "k" and plotting the results.
clusterer = KMeans(n_clusters = 5,
                   random_state=10,
                   n_init = 10)

cluster_labels = clusterer.fit_predict(small_seg)
print(cluster_labels)
```

Silhouette Method

```python
import matplotlib.cm as cm
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# Define the range of cluster numbers to evaluate
range_n_clusters = [2, 3, 4, 5, 6]

for n_clusters in range_n_clusters:  # Iterate through desired cluster numbers
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(small_seg) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility
    clusterer = KMeans(n_clusters=n_clusters, random_state=10, n_init=10)
    cluster_labels = clusterer.fit_predict(small_seg[['Total Revenue_Standardized', 'Avg_Unit_Cost_Standardized']]) # Select the features for clustering

    # The silhouette_score gives the average value for all the samples
    silhouette_avg = silhouette_score(small_seg[['Total Revenue_Standardized', 'Avg_Unit_Cost_Standardized']], cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(small_seg[['Total Revenue_Standardized', 'Avg_Unit_Cost_Standardized']], cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(small_seg['Total Revenue_Standardized'], small_seg['Avg_Unit_Cost_Standardized'], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

plt.show()
```

The value of two cluster in silhouette is method 0.6223660646936714. There is no big variation in silhouette scores of clusters. The value shows data points are clustered better. The best number of clusters would be 5.


Hierarchial Culstering(Dendrogram)

```python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
# Data generation.
np.random.seed(42)

small_den_seg=small_seg.sample(n=10000).reset_index(drop = True)

# Hierarchical clustering from scratch
def hierarchical_clustering(small_den_seg, method='average'):
    dists = pdist(small_den_seg, metric='euclidean')
    linkage_matrix = linkage(dists, method=method)
    return linkage_matrix

# Compute the linkage matrix
linkage_matrix = hierarchical_clustering(small_den_seg)


# Number of clusters we want to extract
k = 5

# Retrieve the clusters
clusters = fcluster(linkage_matrix, k, criterion='maxclust')

# Plot the dendrogram
plt.figure(figsize=(15, 10))
dendrogram(linkage_matrix,truncate_mode='lastp',p=10, leaf_rotation=90, leaf_font_size=12)
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()





```

The optimum number of cluster from dendrogram was 5.


```python
# Perform K-means clustering.
kmeans = KMeans(n_clusters=5,
                random_state=42,
                n_init=10)
small_seg['cluster'] = kmeans.fit_predict(small_seg[['Total Revenue_Standardized','Avg_Unit_Cost_Standardized','Frequency_Standardized','Recency_Standardized','Age_Standardized']])

# Extract centroids.
centroids = kmeans.cluster_centers_

# Plotting the clusters and centroids.
for cluster_idx in range(5):
    cluster_small_seg = small_seg[small_seg['cluster'] == cluster_idx]
    plt.scatter(cluster_small_seg['Total Revenue_Standardized'],
                cluster_small_seg['Avg_Unit_Cost_Standardized'],
                label=f'Cluster {cluster_idx + 1}')
    # Accessing the first two elements of the centroid for x and y coordinates
    plt.scatter(centroids[cluster_idx, 0], centroids[cluster_idx, 1],
                s=200, marker='X', edgecolor='black')

plt.xlabel('Total Revenue_Standardized')
plt.ylabel('Avg_Unit_Cost_Standardized')
plt.xlim(-1.5, 8)
plt.ylim(-1.5, 8)
plt.legend()
plt.show()

# Print final centroids.
print("Final Centroids:", centroids)
```

```python
plt.figure(figsize=(6, 4))
def scatter_plot(data, x, y, hue, title):
    sns.scatterplot(x=x, y=y, hue=hue, data=data, palette='viridis', s=50, edgecolor='k')
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.grid(True)
    plt.legend(title=hue, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

scatter_plot(seg_pca, 'PC1', 'PC2', 'cluster', 'PCA 2D Visualisation with Colour Differentiation')
```

By visually analysing the 2D scatter plot of PCA shows clusters are merged with one another.
